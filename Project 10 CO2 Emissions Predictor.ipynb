# Spliting 

```python
x=er.drop('Value_co2_emissions_kt_by_country',axis=1)
```

```python
y=er['Value_co2_emissions_kt_by_country']
```

```python


from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=30)
```

```python
# scaling
from sklearn.preprocessing import RobustScaler

r=RobustScaler()

train_saclin=r.fit_transform(x_train)

test_saclin=r.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,StackingRegressor,VotingRegressor,BaggingRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score, KFold
```

```python
k=KFold(n_splits=5)
```

```python
list=[LinearRegression(),SVR(),DecisionTreeRegressor(),RandomForestRegressor(),
     KNeighborsRegressor()]
```

```python
for i in list:
    print(i)
    print(cross_val_score(estimator=i,X=train_saclin,y=y_train,cv=k).mean())

```

# VotingRegressor

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.ensemble import StackingRegressor,VotingRegressor,BaggingRegressor,AdaBoostRegressor,GradientBoostingRegressor
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor(max_depth=75)
kn=KNeighborsRegressor(n_neighbors=55)
```

```python
voting_regressor = VotingRegressor(estimators=[('lr', lr), ('kn', kn), ('dt', dt)])
```

```python
voting_regressor.fit(train_saclin,y_train)
```

```python
voting_regressor.score(train_saclin,y_train)
```

```python
voting_regressor.score(test_saclin,y_test)
```

```python
y_pre=voting_regressor.predict(test_saclin)
```

```python
from sklearn.metrics import r2_score,confusion_matrix,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
r2_score(y_test,voting_regressor.predict(test_saclin))
```

```python
mean_absolute_error(y_test,y_pre)
```

```python
mean_squared_error(y_test,y_pre)
```

```python
root_mean_squared_error(y_test,y_pre)
```

# StackingRegressor

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
kn=KNeighborsRegressor(n_neighbors=10)
rf=RandomForestRegressor(n_estimators=200,max_depth=65,min_samples_split=51)
```

```python
sr=StackingRegressor(estimators=[('rf', rf), ('kn', kn), ('dt', dt)],final_estimator=LinearRegression())
```

```python
sr.fit(train_saclin,y_train)
```

```python
sr.score(train_saclin,y_train)
```

```python
sr.score(test_saclin,y_test)
```

```python
y_sr_pre=sr.predict(test_saclin)
```

```python
mean_absolute_error(y_test,y_sr_pre)
```

```python
mean_squared_error(y_test,y_sr_pre)
```

```python
root_mean_squared_error(y_test,y_sr_pre)
```
